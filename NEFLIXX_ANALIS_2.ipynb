```
# Pastikan kolom durasi dalam bentuk numerik
mydata['duration_num'] = pd.to_numeric(mydata['duration_num'], errors='coerce')

# Statistik durasi
durasi_stats = mydata['duration_num'].describe()
print(durasi_stats)
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))

plt.style.use('ggplot')   # membuat style lebih rapi

plt.hist(
    mydata['duration_num'],
    bins=30,
    color='skyblue',       # warna bar
    edgecolor='black'      # garis bar agar lebih tegas
)

plt.title("Distribusi Durasi Konten", fontsize=14, fontweight='bold')
plt.xlabel("Durasi (menit)", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.grid(alpha=0.3)        # grid halus

plt.show()

```

```
mean_dur = mydata['duration_num'].mean()
median_dur = mydata['duration_num'].median()
min_dur = mydata['duration_num'].min()
max_dur = mydata['duration_num'].max()

print(f"Durasi rata-rata: {mean_dur:.2f} menit")
print(f"Durasi median: {median_dur:.2f} menit")
print(f"Durasi minimum: {min_dur} menit")
print(f"Durasi maksimum: {max_dur} menit")

if max_dur > 200:
    print("âš ï¸ Ada durasi lebih dari 200 menit â€” kemungkinan outlier.")
if min_dur < 10:
    print("âš ï¸ Ada durasi kurang dari 10 menit â€” kemungkinan konten pendek/inkonsisten.")
```

# **Perbandingan Antar Kategori**

```
# List rating yang valid (kode klasifikasi umur)
valid_ratings = [
    'TV-MA','TV-14','TV-PG','TV-Y','TV-Y7','TV-Y7-FV',
    'PG-13','R','PG','G','NC-17','NR','UR'
]

# Filter hanya rating yang valid
mydata['rating_clean'] = mydata['rating'].where(mydata['rating'].isin(valid_ratings), pd.NA)

# Cek hasil unique rating setelah cleaning
print(mydata['rating_clean'].unique())
```

setelah di visualisasi ternyata ada format data yang tidak benar pada kolom rating. maka kita menghapus format data tersebut

```
# Hapus kolom lama
mydata = mydata.drop(columns=['rating'])

# Ganti nama rating_clean jadi rating
mydata = mydata.rename(columns={'rating_clean': 'rating'})
```

```
import pandas as pd
import matplotlib.pyplot as plt

# 1. Perbandingan Type vs Rating
type_rating = mydata.groupby(["type", "rating"]).size().unstack().fillna(0)
type_rating.T.plot(kind="bar", figsize=(12,6), stacked=True)
plt.title("Perbandingan Type vs Rating")
plt.ylabel("Jumlah Konten")
plt.xlabel("Rating")
plt.legend(title="Type")
plt.show()

```

**Kesimpulan**

* Mayoritas konten Netflix berada pada rating TV-MA (untuk dewasa) dan TV-14 (remaja ke atas).

* Hal ini menunjukkan bahwa Netflix lebih banyak menyediakan konten untuk penonton dewasa dibanding anak-anak.

* Konten Movie mendominasi di hampir semua kategori rating, sementara TV Show jumlahnya lebih sedikit.

```
# 2. Perbandingan Country vs Type (Top 10 negara)
top_countries = mydata['country'].value_counts().head(10).index
country_type = mydata[mydata['country'].isin(top_countries)].groupby(["country","type"]).size().unstack().fillna(0)
country_type.plot(kind="bar", figsize=(12,6))
plt.title("Perbandingan Type per Negara (Top 10)")
plt.ylabel("Jumlah Konten")
plt.xlabel("Negara")
plt.legend(title="Type")
plt.show()
```

**Kesimpulan**
* Amerika Serikat menjadi penyumbang konten terbanyak

* India berada di posisi kedua dengan jumlah cukup signifikan, sementara negara-negara lain seperti Inggris, Kanada, dan Jepang kontribusinya jauh lebih kecil.

* Sama seperti sebelumnya, Movie lebih dominan dibanding TV Show di hampir semua negara.

```
# 3. Perbandingan Director vs Type (Top 10 sutradara)
top_directors = mydata['director'].value_counts().head(10).index
director_type = mydata[mydata['director'].isin(top_directors)].groupby(["director","type"]).size().unstack().fillna(0)
director_type.plot(kind="bar", figsize=(12,6))
plt.title("Perbandingan Sutradara vs Type (Top 10)")
plt.ylabel("Jumlah Konten")
plt.xlabel("Sutradara")
plt.legend(title="Type")
plt.show()
```

**Kesimpulan**

# **Distribusi Data**

```
import seaborn as sns
import matplotlib.pyplot as plt

# Pastikan sudah ada dataframe: mydata

# Contoh: jika kolom duration masih berbentuk string ("90 min", "1 Season")
# maka buat kolom angka (hanya ambil angka)
mydata['duration_int'] = mydata['duration'].str.extract('(\d+)').astype(float)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribusi Tahun Rilis
sns.histplot(mydata['release_year'], kde=True, ax=axes[0], color="skyblue")
axes[0].set_title("Distribusi Tahun Rilis Konten")

# Distribusi Durasi (menit/season)
sns.histplot(mydata['duration_int'], kde=True, ax=axes[1], color="orange")
axes[1].set_title("Distribusi Durasi Konten")

# Distribusi Rating (pakai countplot karena kategori)
sns.countplot(x='rating', data=mydata, ax=axes[2], palette="Set2")
axes[2].set_title("Distribusi Rating Konten")
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# **Temuan Awal (Insight)**

1. **Dominasi Mutlak Konten Dewasa dan Remaja
Tipe Konten dan Klasifikasi Rating Sangat Terpusat.**

Dari visualisasi perbandingan Type vs Rating, terlihat bahwa TV-MA (khusus dewasa) dan TV-14 (remaja ke atas) adalah kategori rating yang paling dominan, menampung mayoritas konten Movie dan TV Show.

Implikasi: Netflix menargetkan audiens utama mereka pada kelompok usia di atas 14 tahun, menyediakan konten yang lebih sedikit untuk anak-anak (G, TV-Y). Ini menunjukkan bahwa strategi konten Netflix sangat berfokus pada hiburan yang kompleks, matang, atau yang memerlukan pengawasan orang tua.

2.   **Peningkatan Produksi Konten yang Ekstrem Pascatahun 2010
Tren Rilis Tahunan Menunjukkan Ledakan Akuisisi/Produksi.**

Visualisasi tren rilis konten menunjukkan lonjakan yang sangat curam dalam jumlah Movie yang dirilis (dan ditambahkan ke Netflix) setelah tahun 2010, dengan puncaknya terjadi pada 2017-2018.

Implikasi: Ledakan konten ini mengindikasikan masa-masa ekspansi paling agresif Netflix, baik dalam hal akuisisi lisensi dari studio luar maupun peningkatan produksi konten orisinal mereka sendiri. Peningkatan ini didominasi oleh Movie, menunjukkan bahwa Netflix memprioritaskan penambahan film tunggal daripada serial TV Show panjang.

3.   **Durasi Movie Terstandarisasi Meskipun Dominan
Mayoritas Movie Berada di Kisaran Durasi Standar.**

Analisis distribusi durasi Movie menunjukkan pola distribusi yang jelas terpusat antara 80 hingga 120 menit, dengan median tepat di 97 menit.

Implikasi: Meskipun jumlah Movie sangat dominan, durasi konten tersebut sangat terstandarisasi di sekitar durasi film layar lebar tradisional (sekitar 1,5 hingga 2 jam). Hal ini berbeda dengan TV Show di mana 1 Season merupakan durasi yang paling umum, yang sering kali berarti mini-series atau serial yang cepat dihentikan/dibatalkan.


4.   **Ketergantungan Kuat pada Konten Amerika Serikat dan India
Ketersediaan Konten Sangat Terpaku pada Dua Negara.**

Visualisasi perbandingan Country vs Type (Top 10) menegaskan dominasi United States (AS) sebagai sumber konten, tetapi diikuti oleh India dengan jumlah yang signifikan. Kontribusi dari negara-negara lain seperti Inggris, Kanada, dan Jepang jauh lebih kecil.

Implikasi: Meskipun Netflix adalah platform global, koleksi kontennya sangat bergantung pada dua pasar besar: Hollywood (AS) dan industri Bollywood/regional (India). Strategi konten ini menunjukkan bahwa pasar utama untuk akuisisi atau produksi konten Netflix masih berpusat pada industri film berbahasa Inggris dan India.
